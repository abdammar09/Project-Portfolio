### Data Manipulation and Calculations

In the electrical case the charging circuit of a capacitor in series with a resistor has been proven to operate on the principles of the following mathematical equation: <p style="text-align: center;">$$V_t = V_0(1 - e^{\frac{-t}{RC}})$$</p>
Subsequently, leveraging this mathematical relationship in this experiment, which is targeted towards the building and charging of an equivalent hydraulic RC circuit yields the following expression for the hydraulic potential at time t:<p style="text-align: center;">$$V_{H_{t}} = V_{H_{0}}(1 - e^{\frac{-t}{RC}})$$</p>

From the mathematical definition of the hydraulic potential($$V_H = \rho g \Delta h$$), the above expression simplifies to;<p style="text-align: center;">$$\Delta h_t = \Delta h_0(1 - e^{\frac{-t}{R_H C_H}})$$</p>
whose hydraulic time constant is given by the expression; $$\tau = R_H C_H$$
