#### Similarities and Differences between Electric and Hydraulic Circuits

$>$ For both electrical and hydraulic circuits it has been experimentally proven, a non-linear relationship exists between the resistance of a linear circuit element and its effective diameter with the hydraulic resistance having a greater degree of dependance in comparison to the electrical case.
<br>
| Electric | Hydraulic |
| ---------| ----------- |
| The circuit is complete(closed loop) | The circuit is incomplete |
| The movement of electrons form the electric current | The physical displacement of water forms the hydraulic current |
| The intrinsic material property influencing the resistance is termed the resistivity of the wire(channel) and is of the order ranging from the scale of micro to nano | The material property impacting the hydraulic resistance is inherent to the fluid called the coefficient of viscosity as well as the adsorptive(adhesive) property to the inner walls of the capillary tubes(which wasn't measured in this experiment and assumed to be negligible) |
